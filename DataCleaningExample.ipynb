```python
with open("DataDump/Bookie_MLB_2021-04-06.json") as f:
    data = json.load(f)
games_MLB = [k["teams"] for k in data["data"]]
[j["sites"][0] for j in data["data"]]
sites_MLB = [k for k in [j["sites"] for j in data["data"]]]
AllGames_MLB = []
for game in sites_MLB:
    currGameSites = []
    currGameOdds = []
    for i in range(0, len(game)):
        currGameSites.append(game[i]["site_key"])
        currGameOdds.append(game[i]["odds"]["spreads"]["odds"])
    AllGames_MLB.append([currGameSites, currGameOdds])

BettingDict_MLB_Bookie = {}
count = 0
for game in AllGames_MLB:
    currDict = {}
    for i in range(0, len(game[0])):
        currDict[game[0][i]] = game[1][i]
    currDict = dict(sorted(currDict.items(), key = lambda kv: kv[0]))
    currDict["Team_One"] = games_MLB[count][0]
    currDict["Team_Two"] = games_MLB[count][1]
    BettingDict_MLB_Bookie[count] = currDict
    count = count + 1
MLB_Bookie_Frame = pd.DataFrame(BettingDict_MLB_Bookie).transpose()
MLB_Bookie_Frame
#data
```

```python
with open("DataDump/VersusSports_MLB_2021-04-06.json") as f:
    data = json.load(f)
VersusDict_MLB = {}
count = 0
for game in range(0, len(data["game"])):
    currDict = {}
    currDict["team_one"] = data["game"][game]["team1"]
    currDict["team_two"] = data["game"][game]["team2"]
    currDict["date"] = data["game"][game]["gameDate"]
    currDict["team1WinProb"] = data["game"][game]["team1WinProbability"]*.1
    currDict["team2WinProb"] = data["game"][game]["team2WinProbability"]*.1
    VersusDict_MLB[count] = currDict
    count = count + 1
VersusDict_MLB
Versus_MLB_Frame = pd.DataFrame(VersusDict_MLB).transpose()
Versus_MLB_Frame
#1617729000
```

```python
with open("DataDump/VersusSports_NHL_2021-04-06.json") as f:
    data = json.load(f)
VersusDict_NHL = {}
count = 0
for game in range(0, len(data["game"])):
    currDict = {}
    currDict["team_one"] = data["game"][game]["team1"]
    currDict["team_two"] = data["game"][game]["team2"]
    currDict["date"] = data["game"][game]["gameDate"]
    currDict["team1WinProb"] = data["game"][game]["team1WinProbability"]*.1
    currDict["team2WinProb"] = data["game"][game]["team2WinProbability"]*.1
    VersusDict_NHL[count] = currDict
    count = count + 1
VersusDict_NHL
Versus_NHL_Frame = pd.DataFrame(VersusDict_NHL).transpose()
Versus_NHL_Frame
```

```python
MLB_Bookie_Frame["team_one"] = "fill"
MLB_Bookie_Frame["team_two"] = "fill"
for i in range(0,14):
    if MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0] == "San" or MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0] == "St." or MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0] == "Tampa":
        MLB_Bookie_Frame["team_one"][i] = MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0] + " " + MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[1]
    elif MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0] == "Los":
        MLB_Bookie_Frame["team_one"][i] = "LA " + MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[2]
    elif MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0] == "New":
        MLB_Bookie_Frame["team_one"][i] = "NY " + MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[2]
    else:
        MLB_Bookie_Frame["team_one"][i] = MLB_Bookie_Frame["Team_One"].loc[i].split(" ")[0]
        
    if MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0] == "San" or MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0] == "St." or MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0] == "Tampa":
        MLB_Bookie_Frame["team_two"][i] = MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0] + " " + MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[1]
    elif MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0] == "Los":
        MLB_Bookie_Frame["team_two"][i] = "LA " + MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[2]
    elif MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0] == "New":
        MLB_Bookie_Frame["team_two"][i] = "NY " + MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[2]
    else:
        MLB_Bookie_Frame["team_two"][i] = MLB_Bookie_Frame["Team_Two"].loc[i].split(" ")[0]
```

```python
MLB_Frame = pd.merge(MLB_Bookie_Frame, Versus_MLB_Frame, on=["team_one", "team_two"])
MLB_Frame = MLB_Frame.drop("team_one", axis=1)
MLB_Frame = MLB_Frame.drop("team_two", axis=1)
MLB_Frame
```
